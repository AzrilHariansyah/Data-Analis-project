```python
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm

# Baca data dari Excel
df = pd.read_excel('coba yang ini.xlsx')

# Buat kolom tanggal bulanan mulai 2019-01-01
df['tanggal'] = pd.date_range(start='2019-01-01', periods=len(df), freq='MS')

# Pastikan data sudah berurutan berdasarkan tanggal
df = df.sort_values('tanggal')

# Data seri waktu: nilai ekspor
data = df['nilai ekspor']

# Auto ARIMA (fit model)
model = pm.auto_arima(data, seasonal=True, m=12,  # m=12 untuk data bulanan
                      trace=True, error_action='ignore', suppress_warnings=True)

print(model.summary())

# Prediksi 12 bulan ke depan
n_periods = 12
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Buat index tanggal untuk prediksi
forecast_index = pd.date_range(df['tanggal'].iloc[-1] + pd.offsets.MonthBegin(), periods=n_periods, freq='MS')

# Plot data asli dan prediksi
plt.figure(figsize=(10,5))
plt.plot(df['tanggal'], data, label='Data Asli')
plt.plot(forecast_index, forecast, label='Forecast ARIMA', color='red')
plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.3, label='Confidence Interval')
plt.title('Forecast Nilai Ekspor dengan Auto ARIMA')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Ekspor')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Prediksi 10 bulan (Maret sampai Desember 2025)
forecast_index = pd.date_range(start='2025-03-01', end='2025-12-01', freq='MS')
n_periods = len(forecast_index)

forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Plot
plt.figure(figsize=(10,5))
plt.plot(df['tanggal'], data, label='Data Asli')
plt.plot(forecast_index, forecast, label='Forecast ARIMA', color='red')
plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.3, label='Confidence Interval')
plt.title('Forecast Nilai Ekspor dari Maret sampai Desember 2025')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Ekspor')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm

# 1. Baca data
df = pd.read_excel('coba yang ini.xlsx')

# 2. Buat kolom tanggal bulanan mulai 2019-01-01
df['tanggal'] = pd.date_range(start='2019-01-01', periods=len(df), freq='MS')
df = df.sort_values('tanggal')

data = df['nilai ekspor']

# 3. Fit model auto ARIMA
model = pm.auto_arima(data, seasonal=True, m=12,
                      trace=True, error_action='ignore', suppress_warnings=True)

# 4. Prediksi in-sample (fit model pada data training)
in_sample_pred = model.predict_in_sample()

# 5. Prediksi out-of-sample (forecast) dari Maret sampai Desember 2025
forecast_index = pd.date_range(start='2025-03-01', end='2025-12-01', freq='MS')
n_periods = len(forecast_index)
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# 6. Plot data aktual, hasil fit model, dan forecast
plt.figure(figsize=(12,6))

# Data asli
plt.plot(df['tanggal'], data, label='Data Aktual')

# Hasil fit model ARIMA (in-sample prediction)
plt.plot(df['tanggal'], in_sample_pred, label='Fitted ARIMA', color='orange')

# Forecast out-of-sample
plt.plot(forecast_index, forecast, label='Forecast ARIMA', color='red')

# Confidence interval forecast
plt.fill_between(forecast_index, conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.3, label='Confidence Interval')

plt.title('Data Aktual, Fitted Model ARIMA, dan Forecast')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Ekspor')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
