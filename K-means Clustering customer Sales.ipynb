```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from IPython.display import display

# Load data
df = pd.read_csv('ecommerce_sales_analysis.csv')

# Pilih fitur yang akan digunakan untuk clustering
features = ['price', 'review_score', 'review_count'] + [f'sales_month_{i}' for i in range(1,13)]

# Buat subset data untuk clustering, drop baris dengan nilai kosong
X = df[features].dropna()

# Scaling data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Tentukan jumlah cluster (misal 3)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# Tambahkan hasil cluster ke dataframe asli (sesuai index)
df.loc[X.index, 'Cluster'] = kmeans.labels_

# Tampilkan 20 data dengan cluster
display(df.head(20))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load data (kalau belum di-load)
df = pd.read_csv('ecommerce_sales_analysis.csv')

# Fitur yang digunakan clustering
features = ['price', 'review_score', 'review_count'] + [f'sales_month_{i}' for i in range(1,13)]

# Pastikan kolom Cluster sudah ada, kalau belum, jalankan clustering dulu
if 'Cluster' not in df.columns:
    from sklearn.cluster import KMeans
    X = df[features].dropna()
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans.fit(X_scaled)
    df.loc[X.index, 'Cluster'] = kmeans.labels_

# Ambil data yang lengkap dan cluster
df_clustered = df.dropna(subset=features + ['Cluster'])
X = df_clustered[features]
clusters = df_clustered['Cluster'].astype(int)

# Scaling fitur untuk PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA ke 2 komponen
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Buat dataframe hasil PCA dan cluster
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = clusters.values

# Plot hasil clustering dengan PCA
plt.figure(figsize=(10,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set1', s=60)
plt.title('Visualisasi Hasil Clustering K-Means (PCA 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()

# Analisis rata-rata tiap fitur per cluster
cluster_summary = df_clustered.groupby('Cluster')[features].mean().round(2)
display(cluster_summary)
```

```python
# Mapping cluster ke kategori berdasarkan analisis cluster_summary
cluster_mapping = {
    0: 'Medium Priced, Stable Sales',
    1: 'High Priced, Premium Products',
    2: 'Low Priced, Variable Sales'
}

# Pastikan kolom Cluster ada dan bertipe integer
df['Cluster'] = df['Cluster'].astype(int)

# Buat kolom kategori baru berdasarkan mapping cluster
df['Cluster_Category'] = df['Cluster'].map(cluster_mapping)

# Tampilkan 20 data dengan kategori baru
display(df[['product_id', 'price', 'review_score', 'review_count', 'Cluster', 'Cluster_Category']].head(20))
```
